```python
import os
import json
from pydantic import TypeAdapter
from tribulnation.catalogue import Asset, Platform, Network, Catalogue, load
from mexc import MEXC
from mexc.core import ApiError

client = MEXC.new('', '')
```

```python
catalogue = load.all('data')
```

```python
instruments = catalogue.perpetual_instruments['mexc']
nope = []
for i, id in enumerate(instruments):
  print(f'[{i+1}/{len(instruments)}] {id}')
  try:
    await client.futures.contract_info(id)
  except ApiError as e:
    print('>', e)
    nope.append(id)
```

```python
instruments = catalogue.spot_instruments['mexc']
for i, id in enumerate(instruments):
  print(f'[{i+1}/{len(instruments)}] {id}')
  await client.spot.exchange_info(id)
```
